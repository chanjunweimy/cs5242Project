```python
import pandas as pd
import numpy as np

pred = np.zeros((3732,9))
score = np.zeros((9))

for feature in range(9):
    pred[:,feature] = np.squeeze(pd.read_csv('features/feature_{}.csv'.format(feature)).values)
    score[feature] = pd.read_csv('features/score_{}.csv'.format(feature)).columns[0]
```

```python
# All prediction results

pred = pd.DataFrame(data=pred)
pred = pred.rename(index=lambda x: 'test_' + str(x), columns=lambda x: 'feature' + str(x))
pred
```

```python
# Individual feature AUC scores

score = pd.DataFrame(data=score)
score = score.rename(index=lambda x: 'feature_' + str(x), columns=lambda x: 'AUC Score')
score
```

```python
# List all predictions compared to label

ytest = pd.read_csv('features/ytest.csv').values

from sklearn import metrics

pred_ = np.array(pred)
pred['Label'] = ytest
pred
# pred.to_csv("consolidated.csv")
```

```python
# Process pandas to obtain best and relu from features

def best_feature(pred,ytest):    
    score2 = pd.DataFrame(data=ytest)
    score2 = score2.rename(columns={0: "ytest"})

    pred2 = pred
#     pred2 = pred2.drop(columns = ['Label'])
    score2['Max'] = pred2.max(axis=1).values
    score2['Min'] = pred2.min(axis=1).values
    score2['Max_dist'] = 1 - score2['Max']
    score2['Min_dist'] = score2['Min']
    score2['Average'] = pred2.mean(axis=1).values
    score2['Best feature'] = np.where(score2['Max_dist']<=score2['Min_dist'],score2['Max'],score2['Min'])
    score2['ReLu'] = np.where(score2['Max_dist']<=score2['Min_dist'],1,0)
    score2['Label'] = ytest
    score2['Check'] = np.where(score2['ReLu'].values==score2['Label'].values,'True','False')

    score3 = pd.DataFrame(data=score2['Check'])
#     print('Score: ',score3[score3['Check']=='True'].count()/score3.count())
    return score2
```

```python
# Accessing the scores

import itertools

stuff = np.arange(9)

metric = pd.DataFrame(data=np.zeros((511,1)))
metric = metric.rename(columns={0: "Feature"})

subset_list = []
avg = []
best = []
relu = []

for L in range(0, len(stuff)+1):
    for subset in itertools.combinations(stuff, L):
        ypred = pd.DataFrame(data=pred['Label'].values)     
        if subset != ():
            for i in subset:
                ypred['feature_{}'.format(i)] = pred_[:,i]
            ypred = ypred.drop(columns=[0])

            # Average
            avg_pred = ypred.mean(axis=1)

            fpr, tpr, thresholds = metrics.roc_curve(ytest, avg_pred)
            avg_score = metrics.auc(fpr, tpr)

            process_ypred = best_feature(ypred,ytest)

            # Best
            best_ypred = process_ypred['Best feature']

            fpr, tpr, thresholds = metrics.roc_curve(ytest, best_ypred)
            best_score = metrics.auc(fpr, tpr)

            # ReLu
            ReLu_ypred = process_ypred['ReLu']

            fpr, tpr, thresholds = metrics.roc_curve(ytest, ReLu_ypred)
            relu_score = metrics.auc(fpr, tpr)
            
            # Add to list
            subset_list.append(subset)
            avg.append(avg_score)
            best.append(best_score)
            relu.append(relu_score)

            
metric['Feature'] = subset_list
metric['avg AUC'] = avg
metric['best AUC'] = best
metric['relu AUC'] = relu

# metric = metric.sort_values(by=['AUC Score'], ascending=False)
metric
```

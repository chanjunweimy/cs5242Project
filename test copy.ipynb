```python
import numpy as np

def load_dataframe(id):
    train_data = np.load("train/train/{}.npy".format(id))
    return pd.DataFrame(data=train_data)
```

```python
def load_test_dataframe(id):
    test_data = np.load("test/test/{}.npy".format(id))
    return pd.DataFrame(data=test_data)
```

```python
def get_missing_data(df):
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return missing_data
```

```python
def clear_missing_data(df, missing_indices):
    df = df.drop(missing_indices,1)
    # for col in df_temp.columns:
    #    df = df.drop(df.loc[df[col].isnull()].index)
    count = df.isnull().sum().max() #just checking that there's no missing data missing...
    if count > 0:
        print(count)
        return pd.DataFrame()
    return df
```

```python
dataframes = []
for id in df_train['Id']:
    dfi = load_dataframe(id)
    dataframes.append(dfi.values)
dataframes
```

```python
np.save('allData.npy', np.array(dataframes))
```

```python
def pad_data(dfs):
    data = []
    for i in range(len(dfs)):
        df = dfs[i]
        diff = 1000 - df.shape[0]
        if diff > 0:
            halfDiff = int(diff / 2)
            df = np.pad(df, [(halfDiff, halfDiff + diff % 2), (0,0)], 'constant')
        data.append(df)
    data = np.stack(data)
    return data
```
